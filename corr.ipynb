```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_and_sav_corr(df, timeframe):
    fig, ax = plt.subplots(figsize=(8,8))
    sns.heatmap(df.corr(method='pearson'), annot=True, fmt='.4f', 
                cmap=plt.get_cmap('coolwarm'), cbar=False, ax=ax)
    ax.set_yticklabels(ax.get_yticklabels(), rotation="horizontal")
    plt.savefig('images/corr'+str(timeframe)+'.png', bbox_inches='tight', pad_inches=0.0, dpi = 300, format = 'jpg')
    plt.savefig('images/corr'+str(timeframe)+'.pdf', bbox_inches='tight', pad_inches=0.0, dpi = 300, format = 'pdf')
    plt.show()
```

### Day

```python
day = pd.read_csv("data/full_day.csv").corr()
plot_and_sav_corr(day, "day")
```

### Hour

```python
hour = pd.read_csv("data/full_hour.csv").corr()
plot_and_sav_corr(hour, "hour")
```

### Minute

```python
minute = pd.read_csv("data/full_minute.csv").corr()
plot_and_sav_corr(minute, "minute")
```
